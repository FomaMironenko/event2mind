```python
def replace_excess(source):
    f = 0
    while(f != -1):
        f = source.find("<a", f)
        if(f == -1):
            break
        r = source.find("a>", f)
        source = source[:f] + source[r+2:]
    f = 0
    while(f != -1):
        f = source.find("</b", f)
        if(f == -1):
            break
        r = source.find("b>", f)
        source = source[:f] + source[r+2:]
    f = 0
    while(f != -1):
        f = source.find("<b", f)
        if(f == -1):
            break
        r = source.find("b>", f)
        source = source[:f] + source[r+2:]
    return source

def get_text(URL):
    page = requests.get(URL)
    txt = BeautifulSoup(page.text, "html.parser")
    all_sent = txt.find_all("p")
    all_text = ""
    for sent in all_sent[:-5]:
        all_text += str(sent).replace("<p>", "").replace("</p>", "")
    return replace_excess(all_text)
```
