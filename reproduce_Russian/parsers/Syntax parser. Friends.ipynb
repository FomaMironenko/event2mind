```python
import sys
try:
    sys.path.insert(0, "/usr/lib/python3.7/site-packages")
except FileNotFoundError:
    pass

import requests
import pandas as pd
import numpy as nm
from bs4 import BeautifulSoup
```

```python
def get_text(URL):
    page = requests.get(URL)
    txt = BeautifulSoup(page.text, "html.parser")
    all_sent = txt.find_all("p")
    all_text = ""
    for sent in all_sent[:-5]:
        all_text += str(sent.get_text()).replace("<p>", "").replace("</p>", "")
    return all_text
```
